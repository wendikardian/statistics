```python
import seaborn as sns
import numpy as np
from scipy.stats import mode
import matplotlib.pyplot as plt

data = [82, 69, 82, 90, 75, 73, 82, 80, 65, 79, 74, 94, 89, 90, 76, 92, 79, 74, 75, 91, 90, 84, 81, 69, 89, 77, 90, 81, 86, 83]

sns.histplot(data, kde=False, color="skyblue")

mean = np.mean(data)
median = np.median(data)
mode = mode(data)[0][0]

sns.histplot(data, kde=False, color="skyblue")
plt.axvline(x=mean, color="red", linestyle="--")
plt.axvline(x=median, color="green", linestyle="--")
plt.axvline(x=mode, color="orange", linestyle="--")
```
